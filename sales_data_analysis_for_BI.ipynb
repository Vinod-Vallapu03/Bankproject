```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv")

# Rename columns (use Rank as Sales, State as Product)
data.rename(columns={'Rank': 'Sales', 'State': 'Product'}, inplace=True)

# Top 5 products
top_products = data.nlargest(5, "Sales")

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x="Product", y="Sales", data=top_products, palette="viridis")
plt.title("Top 5 Products by Sales")
plt.show()

# Insight
print("Highest sales from:", top_products.iloc[0]['Product'])
```
